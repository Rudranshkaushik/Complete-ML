```python
df['rating'] = df['rating'].apply(lambda x:0 if x<3 else 1)
```

```python
df['reviewText']= df['reviewText'].str.lower()
```

```python
import re 
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
from bs4 import BeautifulSoup
```

```python
df['reviewText']=df['reviewText'].apply(lambda x:re.sub('[^a-z A-z 0-9-]+', '',x))
## Remove the stopswords
```

```python
df['reviewText']=df['reviewText'].apply(lambda x:" ".join([y for y in x.split() if y not in stopwords.words('english')]))
## Remove url 
```

```python
df['reviewText']=df['reviewText'].apply(lambda x: re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '' , str(x)))
## Remove html tags
```

```python

df['reviewText'] = df['reviewText'].astype(str).apply(
    lambda x: BeautifulSoup(x, 'html.parser').get_text()
)
df['reviewText'] = df['reviewText'].apply(lambda x: " ".join(x.split()))
```

```python
df['reviewText']=df['reviewText'].apply(lambda x: " ".join(x.split()))
df.head()
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemm = WordNetLemmatizer()
```

```python
def lemm_words(text):
    return " ".join([lemm.lemmatize(word) for word in text.split()])
```

```python
df['reviewText'] = df['reviewText'].apply(lambda x: lemm_words(x))
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(df['reviewText'],df['rating'],test_size=0.20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

bow = CountVectorizer()
x_train_bow = bow.fit_transform(x_train).toarray()
x_test_bow = bow.fit_transform(x_test).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
x_train_tfidf = tfidf.fit_transform(x_train).toarray()
x_test_tfidf = tfidf.fit_transform(x_test).toarray()
```

```python
import numpy as np
from gensim.utils import simple_preprocess
```

```python
import gensim.downloader as api

wv = api.load('word2vec-google-news-300')
```

```python

x_train_tokens = [simple_preprocess(doc) for doc in x_train]
x_test_tokens  = [simple_preprocess(doc) for doc in x_test]
```

```python

DIM = wv.vector_size  # 300 for Google News

def doc_vector_avg(tokens, keyed_vectors, dim):
    vecs = [keyed_vectors[w] for w in tokens if w in keyed_vectors]
    if not vecs:
        return np.zeros(dim, dtype=np.float32)
    return np.mean(vecs, axis=0)
```

```python

x_train_w2v = np.vstack([doc_vector_avg(t, wv, DIM) for t in x_train_tokens])
x_test_w2v  = np.vstack([doc_vector_avg(t, wv, DIM) for t in x_test_tokens])
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model_bow=GaussianNB().fit(x_train_bow,y_train)
nb_model_tfidf=GaussianNB().fit(x_train_tfidf,y_train)
nb_model_w2v = GaussianNB().fit(x_train_w2v,y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
y_pred_w2v = nb_model_w2v.predict(x_test_w2v)
```

```python
confusion_matrix(y_test, y_pred_w2v)
```

```python
print("w2v accuracy: ",accuracy_score(y_test,y_pred_w2v))
```
